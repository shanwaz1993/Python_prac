```python
import seaborn as sns
import pandas as pd

planets = sns.load_dataset('planets')
planets.head()
planets.describe()
```

```python
planets['decade']=10*(planets['year']//10)
planets.head()
planets.drop('year',1).head()  #drop the column not required ...year note :categorieze as decade
```

```python
planets.query('decade<2010 & mass>10')['mass'].max()  #query 
```
