```python
import matplotlib as mlt
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.plot(x,np.sin(x-0),'--',color='K')
plt.plot(x,np.sin(x-1),'--',color='b')         #b--blue 
plt.plot(x,np.sin(x-1),'--',color='0.75')      #Grayscale between 0 and 1
plt.plot(x,np.sin(x-2),'--',color='#FFDD44')  #Hex code (RRGGBB from 00 to FF)
plt.plot(x,np.sin(x-3),'--',color=(1.0,0.2,0.3))  #RGB tuple, values 0 and 1
```

```python
plt.plot(x,x+4,linestyle='-')     # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.')  # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
plt.plot(x,x+1,linestyle='solid')     # solid
plt.plot(x, x + 2, linestyle='dashed') # dashed
plt.plot(x, x + 3, linestyle='dashdot')  # dashdot
plt.plot(x, x + 4, linestyle='dotted'); # dotted

```

```python
plt.plot(x,x+1,'-k')     
plt.plot(x, x + 2,'--c')
plt.plot(x, x + 3,'-.k')  
plt.plot(x, x + 4,':r')
```

```python
#Adjusting the Plot: Axes Limits

plt.plot(x, np.sin(x),':r')

b=np.linspace(10,20,100)
plt.plot(b, np.cos(b),'-b')
plt.xlim(-1, 20)
plt.ylim(-1.5, 1.5)
#plt.show()
```

```python
#plt.axis()

plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
plt.axis('equal')
```

```python
#Labeling Plots

plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
#other method for labelling

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('tight')
plt.legend()  #keeps track of the line style and color..
```

```python
#plt.scatter from plt.plot is that it can be used to create
#scatter plots where the properties of each individual point (size, face color, edge color,
#etc.) can be individually controlled or mapped to data.

rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,cmap='viridis')  #plt.scatter()
plt.colorbar(); # show color scale

#Note--scattter plot is inefficient for large data set  plt.plot should be preferred over plt.scatter for large
#datasets
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T
#print(iris.target)
plt.scatter(features[0], features[1], alpha=0.4,s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
